```python
# Libraries for data manipulation and analysis
import pandas as pd  # Used for data handling and manipulation, especially with dataframes
import numpy as np  # Used for numerical operations, arrays, and matrix manipulations

# Libraries for data visualization
import seaborn as sns  # Used for creating statistical visualizations (e.g., heatmaps, pair plots)
import matplotlib.pyplot as plt  # Used for plotting various types of graphs (line plots, histograms, scatter plots)

# Regression models from scikit-learn (various types of regression models for machine learning)
from sklearn.linear_model import LinearRegression  # Simple linear regression model
from sklearn.linear_model import Ridge  # Linear regression model with L2 regularization (reduces overfitting)
from sklearn import linear_model  # Contains other linear models like Lasso, ElasticNet for additional flexibility
from sklearn.tree import DecisionTreeRegressor  # Decision Tree regression model (for modeling non-linear relationships)
from sklearn.ensemble import RandomForestRegressor  # Random Forest regression model (ensemble of decision trees)
from sklearn.svm import SVR  # Support Vector Regression (SVR) for regression tasks, handles non-linear relationships
from xgboost import XGBRegressor  # XGBoost regressor, an efficient implementation of gradient boosting for regression
from sklearn.neighbors import KNeighborsRegressor  # K-Nearest Neighbors regression model (based on proximity of data points)
from sklearn.ensemble import GradientBoostingRegressor  # Gradient boosting regression model (boosting ensemble method)

# Utility functions for data processing and splitting
from sklearn.model_selection import train_test_split  # Splits data into training and testing sets for model evaluation
from sklearn.preprocessing import LabelEncoder  # Encodes categorical variables into numerical format
from sklearn.preprocessing import MinMaxScaler  # Scales/normalizes data to a specific range (commonly [0, 1])

# Function to evaluate regression model performance
from sklearn.metrics import r2_score  # R-squared metric to evaluate model accuracy (coefficient of determination)

# Ignore warnings for a cleaner output (optional)
import warnings
warnings.filterwarnings('ignore')  # Suppresses warnings from libraries for a clean display (not recommended in debugging)
```

```python
# storing the Dependent Variables in X and Independent Variable in Y
X=data.drop(['price'],axis=1)
y=data['price']
```

```python
y
```

```python
# Splitting the Data into Training set and Testing Set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# Scaling the values to convert the int values to Machine Learning compatible format
mmscaler = MinMaxScaler(feature_range=(0, 1))  # Initialize the MinMaxScaler to scale features to a range of [0, 1]

# Fit and transform the training data
X_train = mmscaler.fit_transform(X_train)  # Scale the training data based on the min and max values of the training set

# Fit and transform the testing data (this is incorrect usage, see explanation below)
X_test = mmscaler.fit_transform(X_test)  # Scale the testing data based on the min and max values of the training set

# Convert the scaled arrays back to DataFrames for easier handling
X_train = pd.DataFrame(X_train)  
X_test = pd.DataFrame(X_test)  
```

```python
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
r2_1 = r2_score(y_test, y_pred)
print("R-squared:", r2_1)
rounded_r2 = round(r2_1, 4) 
print("R-squared (rounded):", rounded_r2*100)
```
